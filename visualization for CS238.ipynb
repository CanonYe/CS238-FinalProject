```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import pandas as pd
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
df=pd.read_csv('./uvapos.csv', sep='\t',header=None)
X = np.asarray(df[0], dtype=np.float64)
Y = np.asarray(df[1], dtype=np.float64)
Z = np.asarray(df[2], dtype=np.float64)

df = pd.read_csv('./targetpos.csv', sep='\t',header=None)
x = np.asarray(df[0], dtype=np.float64)
y = np.asarray(df[1], dtype=np.float64)


fig = plt.figure()
ax = p3.Axes3D(fig)

# create the first plot
point, = ax.plot([X[0]], [Y[0]], [Z[0]], 'o')
point2, = ax.plot([x[0]], [y[0]], '^')
ax.set_xlim([-5, 100])
ax.set_ylim([-5, 100])
ax.set_zlim([-5, 100])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')


# second option - move the point position at every frame
def update_point(n):
    #point.set_data(np.array([[X[n],x[n]], [Y[n],y[n]]]))
    point.set_data(np.array([X[n], Y[n]]))
    point.set_3d_properties(Z[n], 'z')
    point2.set_data(np.array([x[n], y[n]]))
    point2.set_3d_properties(0, 'z')
   # ax.set_title("Time {}".format(n))
    

ani=animation.FuncAnimation(fig, update_point, len(X))
#change view rotation 
#ax.view_init(azim=0, elev=90)
plt.show()
```
